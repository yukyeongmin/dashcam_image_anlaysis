```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import random
from matplotlib.pyplot import figure
import math
```

### 이미지 불러오기 ###

```python
path = 'dashcam1.png'
frame = cv2.imread(path)
```

```python
# reads frames from a camera
#ret, frame = cap.read()

# converting BGR to HSV
hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

# define range of red color in HSV
lower_red = np.array([30,150,50])
upper_red = np.array([255,255,180])

# create a red HSV colour boundary and 
# threshold HSV image
mask = cv2.inRange(hsv, lower_red, upper_red)

# Bitwise-AND mask and original image
res = cv2.bitwise_and(frame,frame, mask= mask)

# Display an original image
#cv2.imshow('Original',frame)

# finds edges in the input image image and
# marks them in the output map edges
edges = cv2.Canny(frame,100,300,L2gradient = True)

# Display edges in a frame
# cv2.imshow('Edges',edges)
```

### 임의의 두 선을 택해 교점 찾기 ###

```python
# line = x1 y1 x2 y2
def line_intersection(line1,line2):
    xdiff= (line1[0][0]-line1[0][2],line2[0][0]-line2[0][2])
    ydiff= (line1[0][1]-line1[0][3],line2[0][1]-line2[0][3])
    
    def det(a,b):
        return a[0]*b[1]-a[1]*b[0]
    
    div = det(xdiff, ydiff)
    
    if div == 0:
        return 2000,2000 # to make continue

    d = (det(line1[0][0:2],line1[0][2:]), det(line2[0][0:2],line2[0][2:]))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    
    return x, y
```

```python
repeat=7

x_repeat=[]
y_repeat=[]

while len(x_repeat)<repeat:
    line1=random.randint(0,lines.shape[0]-1)
    line2=random.randint(0,lines.shape[0]-1)
    
    if line1==line2:
        continue

    x,y=line_intersection(lines[line1],lines[line2])
    
    if x<0 or x>1500:
        continue
    if y<0 or y>800:
        continue
        
    if x<1 or y<1:
        print('0')
        print(lines[line1],lines[line2])
    
    x_repeat.append(x)
    y_repeat.append(y)
```

### 소실점 찾기 ###

```python
def is_contain(point,line, gap):
    
    slope=(line[0][3]-line[0][1])/(line[0][2]-line[0][0])
    
    distance=abs(slope*point[0]-point[1]-slope*line[0][0]+line[0][1])/math.sqrt(slope*slope+1)

    if distance<gap:
        return True
    return False
```

```python
count=np.zeros(repeat)

for index,intersection in enumerate(intersections):
    for line in lines:
        if is_contain(intersection,line, 10):
            count[index]+=1
        # 각 라인이 해당 점을 지나는지 확인
    
```
