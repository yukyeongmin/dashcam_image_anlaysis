### 임의의 두 선을 택해 교점 찾기 ###

```python
# line = x1 y1 x2 y2
def line_intersection(line1,line2):
    xdiff= (line1[0][0]-line1[0][2],line2[0][0]-line2[0][2])
    ydiff= (line1[0][1]-line1[0][3],line2[0][1]-line2[0][3])
    
    def det(a,b):
        return a[0]*b[1]-a[1]*b[0]
    
    div = det(xdiff, ydiff)
    
    if div == 0:
        return 2000,2000 # to make continue

    d = (det(line1[0][0:2],line1[0][2:]), det(line2[0][0:2],line2[0][2:]))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    
    return x, y
```

```python
def find_intersections(repeat,lines):
    x_repeat=[]
    y_repeat=[]

    while len(x_repeat)<repeat:
        line1=random.randint(0,lines.shape[0]-1)
        line2=random.randint(0,lines.shape[0]-1)

        if line1==line2:
            continue

        x,y=line_intersection(lines[line1],lines[line2])

        if x<0 or x>1500:
            continue
        if y<0 or y>800:
            continue

        if x<1 or y<1:
            print('0')
            print(lines[line1],lines[line2])

        x_repeat.append(x)
        y_repeat.append(y)
        
    return x_repeat,y_repeat
```

```python
repeat=7
x_repeat,y_repeat = find_intersections(repeat,lines)
```

### 소실점 찾기 ###

```python
def is_contain(point,line, gap):
    
    slope=(line[0][3]-line[0][1])/(line[0][2]-line[0][0])
    
    distance=abs(slope*point[0]-point[1]-slope*line[0][0]+line[0][1])/math.sqrt(slope*slope+1)

    if distance<gap:
        return True
    return False
```

```python
def find_vanishing_point(repeat,intersections):
    count=np.zeros(repeat)

    for index,intersection in enumerate(intersections):
        for line in lines:
            if is_contain(intersection,line, 10):
                count[index]+=1
            # 각 라인이 해당 점을 지나는지 확인

    for i in range(repeat):
        if count[i]==max(count):
            vanishing=(intersections[:,0][i],intersections[:,1][i])   
        
    return count, vanishing
```

```python
count, vanishing = find_vanishing_point(repeat,intersections)
```

```python
def road(vanishing,edges):
    row, col =edges.shape

    road=np.zeros((row,col))
    road_points=[]
    trapezoid=[]

    v=int(vanishing[0])

    # 사다리꼴을 만들기 위해 위아래 일부점 제외
    for r in range(int(vanishing[1])+100,row-200):
        for c in range(v,v+int(col/2)):
            if edges[r][c]==255:
                road[r][c]=255
                if r==int(vanishing[1])+100 or r==row-200-1 :
                    trapezoid.append([c,r])
                road_points.append([c,r])
                break

    for r in range(int(vanishing[1])+100,row-200):
        for c in range(v,0,-1):
            if edges[r][c]==255:
                road[r][c]=255
                if r==int(vanishing[1])+100 or r==row-200-1 :
                    trapezoid.append([c,r])
                road_points.append([c,r])
                break
                
    return road, road_points, trapezoid
```

```python
road,road_points,trapezoid=road(vanishing,edges) # cv2_grayimage, list, list
road_points=np.array(road_points)
trapezoid= np.array(trapezoid)
```
